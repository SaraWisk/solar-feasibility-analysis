```python
data = pd.read_csv("../data/weatherHistory.csv")
```

```python
data.columns = [col.lower().replace(" ", "_") for col in data.columns]
```

# Question to answer
- What is the distribution of daylight hours in the year 2006?
- What is the distribution of cloud cover in the year 2006?

```python
data["formatted_date"] = pd.to_datetime(data["formatted_date"],utc=True)
```

```python
data["formatted_date"].dt.year
```

```python
df_2006 = data[data["formatted_date"].dt.year==2006]
```

```python
df_2006 = df_2006.sort_values(by="formatted_date").reset_index(drop=True)
```

```python
df_2006
```

```python
df_2006.info()
```

```python
df_2006["summary"].value_counts()
```

```python
df_2006["visibility_(km)"].plot()
```

```python
df_2006_sub = df_2006[df_2006["formatted_date"].dt.month==1]
```

```python
plt.plot(df_2006_sub["formatted_date"],df_2006_sub["visibility_(km)"])
plt.tight_layout()
plt.show()
```

```python
df_2006_month = df_2006.groupby(by=[df_2006["formatted_date"].dt.month]).mean()
df_2006_month
```

```python
df_2006_month.insert(0,"Months",['Jan','Feb','Mar','Apr','May','June','July','Aug','Sep','Oct','Nov','Dec'])
```

```python
df_2006_month
```

```python
fig = plt.figure(figsize =(10, 5))
plt.bar(df_2006_month["Months"],df_2006_month["visibility_(km)"])
plt.ylabel("Average Visibility (km)")
plt.title("Average Visibility (km) per Month in 2006")
```
